#### 1. DISTRIBUCION DE VUELOS RESPECTO AL AÑO

¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

```python
# Para obtener esta información necesitamos las columnas:
# Flights Booked --> num total vuelos AL MES
# Year --> registro de las actividades del vuelo de ese cliente, en ese año

# Recuperamos los datos únicos de la columna 'Year', para saber cuántos años tenemos
print(df_customers['Year'].unique()) # --> 2017, 2018

# Definimos dos df por cada año, para facilitar el trabajo
df_2017 = df_customers [df_customers['Year'] == 2017]
df_2018 = df_customers [df_customers['Year'] == 2018]
```

```python
# Vemos cuántas reservas de vuelos tenemos por mes en cada año:

print('Distribución de vuelos por mes en 2017')
df_fb_17 = df_2017.groupby('Month')['Flights Booked'].sum().reset_index() # agrupamos por num de reservas al mes, lo convertimos en DF
display(df_fb_17)
print('.'*100)

print('Distribución de vuelos por mes en 2018')
df_fb_18 = df_2018.groupby('Month')['Flights Booked'].sum().reset_index() # agrupamos por num de reservas al mes, lo convertimos en DF
display(df_fb_18)
print('.'*100)
```

==> con una `VISUALIZACIÓN` podremos verlo de forma muy clara:

```python
# --> 2017
# En este caso, al haber una variable temporal, usaremos un LINEPLOT :
valores_tiempo_17 = list(df_fb_17['Month'])
valores_reservas_17 = list(df_fb_17['Flights Booked'])
sns.lineplot( x = valores_tiempo_17,
              y = valores_reservas_17) 

# Podemos superponer un gráfico tipo BARPLOT, 
# ya que la columna 'Month' la trataremos como una columna categórica:
plt.bar(data=df_fb_17, 
        x= 'Month', 
        height= 'Flights Booked',
        color = 'green',
        alpha = 0.5)
              
plt.title("Flights Booked in 2017")
plt.grid()  # activamos la cuadrícula para ver mejor los valores
plt.xlim(0,13) # límite eje x
plt.xticks(ticks=list(range(1,13))) # valor ticks hay que pasarle una lista --> que aparezcan todos los meses
plt.xlabel("Month") # ponemos nombre al eje x
plt.ylim(0, 110000);  # límite eje y
```

```python
# --> 2018
# LINEPLOT
valores_tiempo_18 = list(df_fb_18['Month'])
valores_reservas_18 = list(df_fb_18['Flights Booked'])
sns.lineplot( x = valores_tiempo_18,
              y = valores_reservas_18)
#  BARPLOT 
plt.bar(data=df_fb_18, 
        x= 'Month', 
        height= 'Flights Booked',
        color = 'green',
        alpha = 0.5)
plt.title("Flights Booked in 2018")
plt.grid() 
plt.xlim(0,13) 
plt.xticks(ticks=list(range(1,13))) 
plt.xlabel("Month")
plt.ylim(0, 110000); 
```

--> Se observa que ambos años son muy parecidos, siendo los meses de mayores y menores reservas los mismos.  
--> Para compararlos, uniremos los gráficos de línea: 

```python
# Gráficos lineales
sns.lineplot( x = valores_tiempo_17,
              y = valores_reservas_17,
              label = 'Year 2017')       # incorporamos la etiqueta
sns.lineplot( x = valores_tiempo_18,
              y = valores_reservas_18,
              label = 'Year 2018')       # incorporamos la etiqueta

plt.title("Booked Flights per Month")
plt.xlim(1,12)  # límite eje x
plt.ylim(40000, 110000) # límite eje y
plt.xlabel('Month')  # etiqueta eje x
plt.xticks(ticks=list(range(1,13))) # que aparezcan todos los meses
plt.legend(loc = 'lower right');
```

--> La distribución es la misma, aunque los valores no lo sean: Se ve claramente que la distribución por mes al año tiene una tendencia clara.

#### 3. DISTRIBUCION CLIENTES POR PROVINCIA

¿Cuál es la distribución de los clientes por provincia o estado?

```python
# Para obtener esta información necesitamos la columna:
# Province (obj)

# Vamos a filtrar el df para obtener el Loyalty Number sin repetir : sabemos que teníamos 16737 clientes
# (o podríamos usar directamente el df limpio "Customer Loyalty History")
df_ln = df_customers[['Loyalty Number', 'Province']]
df_unicos = df_ln.drop_duplicates() # df con la provincia de cada cliente
print(f'Tenemos {len(df_unicos)} clientes únicos.') # comprobamos

# Calculamos el recuento por cada provincia:
df_provincia = df_unicos.groupby('Province').count().reset_index()
df_provincia.columns = ['Province','Customers']
df_provincia.sort_values(by='Customers', ascending = False, inplace=True)
df_provincia.reset_index(drop=True, inplace=True)
        # añadimos una columna al df con los %
porc = np.array(df_provincia['Customers']) # de serie a array (int)
escalar = 167.37 
porc = np.round(porc / escalar, 2)
df_provincia['%'] = porc
display(df_provincia)

# Graficamos
# Usaremos un COUNTPLOT, ya que tenemos demasiadas variables como para usar un PIEPLOT:
orden_province = list(df_provincia['Province']) # lista en orden de M a m
sns.countplot(data=df_unicos, 
              x = 'Province',
              order = orden_province)

plt.title("Customers per Province")
plt.xticks(rotation = 90) # Ponemos los nombres en vertical para que entren enteros
plt.xlabel("Customers") # Ponemos el nombre en español
plt.ylabel("Province") # Ponemos el nombre en español
plt.xticks(fontsize=8);  # Cambiamos el tamaño de las etiquetas del eje x
```

#### 6. CLIENTES SEGÚN ESTADO CIVIL Y GÉNERO

¿Cómo se distribuyen los clientes según su estado civil y género?

```python
# Observamos los datos:
df_gender = df_customers[['Loyalty Number', 'Marital Status', 'Gender']]

m_mujeres = df_gender['Gender'] == 'Female'
m_hombres = df_gender['Gender'] == 'Male'

df_mujeres= df_gender[m_mujeres].drop_duplicates()  # --> 8410
df_mujeres.drop('Gender', axis=1, inplace=True)   # quitamos ya la columna de género, ya no la necesitamos
df_mujeres = df_mujeres.groupby('Marital Status')['Loyalty Number'].count().reset_index() # agrupamos
df_mujeres.columns = ['Marital Status', 'Female Customers']  # renombramos las columnas
df_mujeres.sort_values(by = 'Female Customers', ascending = False, inplace=True) # ordenamos
df_mujeres.reset_index(drop=True, inplace=True)
print('Clientas Mujeres')
display(df_mujeres)

df_hombres= df_gender[m_hombres].drop_duplicates()  # --> 8327 = 16.737   ok
df_hombres.drop('Gender', axis=1, inplace=True)   # quitamos ya la columna de género, ya no la necesitamos
df_hombres = df_hombres.groupby('Marital Status')['Loyalty Number'].count().reset_index() # agrupamos
df_hombres.columns = ['Marital Status', 'Male Customers']  # renombramos las columnas
df_hombres.sort_values(by = 'Male Customers', ascending = False, inplace=True) # ordenamos
df_hombres.reset_index(drop=True, inplace=True)
print('Clientes Hombres')
display(df_hombres)

# Son dos tipos de datos categóricos, por lo que para graficarlos usaremos un COUNTPLOT
sns.countplot(data=df_customers, 
              x='Marital Status', 
              hue = 'Gender') # el género aparecerá por cada estado civil

plt.title('Marital Status - Gender  Customers Distribution')
plt.ylabel('Customers')
plt.xlabel('Marital Status')
plt.xticks(fontsize=10);  # Cambiamos el tamaño de las etiquetas del eje x;
```

--> Vemos que en ambos casos de mujeres y hombres, las diferencias respecto del estado civil son notorias  
--> Por otro lado, tenemos prácticamente el mismo número de hombres y mujeres en cada estado civil.
